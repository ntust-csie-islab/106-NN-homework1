```python
TX = '2'
DX = '0'
train = read_csv('/home/lonely/Desktop/Training Data version3/KDD_withDB_version3_' + TX + 'D' + DX + '.csv', engine='python')
train

a = train.groupby(['Day']).mean()
print train.mean()

features = train.values[:,:-1] #[row:col]
label = train.values[:,-1]
regr = linear_model.LinearRegression()
regr.fit(features, label)
```

```python
def writecsv(time_window, volumn):
    global csvHeaderFlag
    global TX
    global DX
    with open('predict_data0411_0100.csv', 'a') as csvfile:
            fieldnames = ['tollgate', 'time_window', 'direction', 'volume']
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            if(not csvHeaderFlag):
                writer.writeheader()
                csvHeaderFlag = True
            writer.writerow({'tollgate' : TX, 'time_window' : time_window, 'direction': DX, 'volume': volumn})
```

```python
def tempH(i):
    H = ['08','09','10','17','18','19']
    k = 0
    tempH = ''
    tempH_next = ''
    if i > 6:
        i = i % 6
        k = 3
        if i == 0:
            tempH = H[-2]
            tempH_next = H[-1]
    if i == 1 or i == 2:
        tempH = H[0+k]
        tempH_next = H[0+k]
    if i == 3:
        tempH = H[0+k]
        tempH_next = H[1+k]
    if i == 4 or i == 5:
        tempH = H[1+k]
        tempH_next = H[1+k]
    if i == 6:
        tempH = H[1+k]
        tempH_next = H[2+k]
    l = [tempH,tempH_next]
    return l
def tempday(day_i):
    if day_i == 1:
        tempDay = 24
    else:
        tempDay = day_i + 16
    return tempDay
```

```python
def time_windowParsing(day_i, startTime_i):
    H = ['08','09','10','17','18','19']
    M = ['00','20','40']
    d = '2016-10-' + str(tempday(day_i)) # when 2 -> 19, when 1 -> 24
    tempH_variable = tempH(startTime_i)
    #print tempH_variable
    result = '[' + d + ' ' + tempH_variable[0] + ':' + M[(startTime_i - 1) % 3] + ':00,' + d + ' ' + tempH_variable[1] + ':' + M[(startTime_i) % 3] + ':00)'
    return result
#time_windowParsing(7,12)
```

```python
csvHeaderFlag = False
for day_i in range(1,7+1):
    for startTime_i in range(1,12+1):       
        x_predict = np.array([day_i,startTime_i,train.mean()['x1'],train.mean()['x2'],train.mean()['h']])#x1 avg, x2 avg, h avg 
        #print str(day_i) + ' ' + str(startTime_i) + ' ' + str(regr.predict(x_predict))
        writecsv(time_windowParsing(day_i, startTime_i), regr.predict(x_predict))
```
