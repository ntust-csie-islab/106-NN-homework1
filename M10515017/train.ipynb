```python
#preprosessing

table5['starting_time'] = pd.to_datetime(table5['starting_time'])
table5['travel_seq_split'] = table5['travel_seq'].map(lambda x: x.split(';'))
table5['hour'] = table5['starting_time'].dt.hour
table5['week'] = table5['starting_time'].dt.weekday
```

```python
#declare list varible time_100~time_123
for i in range(100,124):
    locals()['time_%d' % (i)] = []
    
seq_sum = []
```

```python
#calculate time of each line
for row in range(len(table5)):
    
    sum=0
    
    for listnum in range(len(table5['travel_seq_split'][row])):
        
        x = table5['travel_seq_split'][row][listnum].split('#')
        sum+=float(x[2])
        locals()['time_%s' % x[0]].append(float(x[2]))
        
    seq_sum.append(sum)

table5['seq_sum'] = seq_sum
```

```python
#Print average time of each line
for i in range(100,124):
    locals()['time_%d' % (i)] = np.array(locals()['time_%d' % (i)])
    print('line %d :%f')%(i, np.average(locals()['time_%d' % (i)]))
```
